```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
v_max = 10
v_min = 1
total_unknown_cells = 100

def power_law_curve(x, p):
    """
    Computes y for a given x and exponent p using:
    y(x) = v_min + (v_max - v_min) * [1 - (x / total_unknown_cells)^p]
    """
    return v_min + (v_max - v_min) * (1 - (x / total_unknown_cells)**p)

# Generate x values
x = np.linspace(0, total_unknown_cells, 500)

# Different p values to visualize the effect on the curve
p_values = [0.1, 0.3, 0.5, 1, 2, 5, 10, 30, 50, 100]

plt.figure(figsize=(10, 6))
for p in p_values:
    y = power_law_curve(x, p)
    plt.plot(x, y, label=f'p = {p}')

plt.xlabel('# unknown cells percentage')
plt.ylabel('v_max')
plt.title('Visualization of the Power-Law Curve for Various p Values')
plt.legend()
plt.grid(True)
plt.show()
```
